```python
# import tensorflow as tf

# IMG_SIZE = 224
# BATCH = 8

# def load_image(path, label):
#     img = tf.io.read_file(path)
#     img = tf.image.decode_jpeg(img, channels=3)
#     img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE))
#     img = img / 255.0
#     return img, label

# def make_dataset(df, shuffle=False):
#     paths = df["abs_path"].values
#     labels = df["category"].values - 1
#     ds = tf.data.Dataset.from_tensor_slices((paths, labels))
#     ds = ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
#     if shuffle:
#         ds = ds.shuffle(20000)
#     ds = ds.batch(BATCH).prefetch(tf.data.AUTOTUNE)
#     return ds

# train_ds = make_dataset(train_df, shuffle=True)
# val_ds = make_dataset(val_df)
# test_ds = make_dataset(test_df)

# print(train_ds)

import tensorflow as tf

IMG_SIZE = 224
BATCH = 32
AUTOTUNE = tf.data.AUTOTUNE
NUM_CLASSES = len(pd.read_csv(f"{ANNO_DIR}/list_category_cloth.txt", sep=r"\s+", header=None, skiprows=2)[0])

def preprocess_image(image):
    image = tf.image.convert_image_dtype(image, tf.float32)  # 0..1
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    return image

def train_augment(image):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, 0.1)
    image = tf.image.random_contrast(image, 0.9, 1.1)
    # add more transforms if you want
    return image

def load_path_label(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = preprocess_image(img)
    return img, label

def make_dataset_from_df(df_subset, shuffle=False):
    paths = df_subset["abs_path"].values.astype(str)
    labels = (df_subset["category"].values - 1).astype('int32')  # convert to 0-index
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(lambda p, l: tf.py_function(func=load_path_label, inp=[p, l], Tout=(tf.float32, tf.int32)),
                num_parallel_calls=AUTOTUNE)
    # tf.py_function loses shape info; set shapes for downstream performance
    def set_shapes(img, lbl):
        img.set_shape([None, None, 3])
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
        img.set_shape([IMG_SIZE, IMG_SIZE, 3])
        lbl.set_shape([])
        return img, lbl
    ds = ds.map(set_shapes, num_parallel_calls=AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(20000)
    ds = ds.map(lambda img, lbl: (train_augment(img), lbl) if shuffle else (img, lbl),
                num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH)
    ds = ds.prefetch(AUTOTUNE)
    return ds

```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers, Model

# base = EfficientNetB0(
#     include_top=False,
#     input_shape=(IMG_SIZE, IMG_SIZE, 3),
#     weights="imagenet"
# )
# base.trainable = False  # Freeze for transfer learning

# inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
# x = base(inputs, training=False)
# x = layers.GlobalAveragePooling2D()(x)
# x = layers.Dropout(0.2)(x)
# outputs = layers.Dense(len(cat_cloth), activation="softmax")(x)

# model = Model(inputs, outputs)

# model.compile(
#     optimizer=tf.keras.optimizers.Adam(1e-3),
#     loss="sparse_categorical_crossentropy",
#     metrics=["accuracy"]
# )

# model.summary()


base = tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)
base.trainable = False

inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.3)(x)
outputs = tf.keras.layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = tf.keras.Model(inputs, outputs)

# model.compile(
#     optimizer=tf.keras.optimizers.Adam(1e-3),
#     loss="sparse_categorical_crossentropy",
#     metrics=["accuracy"]
# )

for layer in base.layers[-50:]:
    layer.trainable = True

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4), #<-- make it 1e-3
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)



# from tensorflow.keras import layers, Model
# from tensorflow.keras.applications import EfficientNetB0

# def build_model(num_classes):
#     base = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
#     base.trainable = False  # initial freeze
#     x = layers.GlobalAveragePooling2D()(base.output)
#     x = layers.Dropout(0.2)(x)
#     out = layers.Dense(num_classes, activation='softmax')(x)
#     model = Model(base.input, out)
#     model.compile(
#         optimizer=tf.keras.optimizers.Adam(1e-3),
#         loss='sparse_categorical_crossentropy',
#         metrics=['accuracy']
#     )
#     return model

# model = build_model(NUM_CLASSES)
# model.summary()

```

```python
ckpt_dir = os.path.join(WORK_DIR, "checkpoints")
os.makedirs(ckpt_dir, exist_ok=True)

checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(ckpt_dir, "model_epoch_{epoch:02d}_valacc_{val_accuracy:.4f}.h5"),
    monitor='val_accuracy',
    verbose=1,
    save_best_only=False,     # saves every epoch; change to True to keep only the best
    save_weights_only=False
)

best_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(ckpt_dir, "model_best.h5"),
    monitor='val_accuracy',
    verbose=1,
    save_best_only=True,
    save_weights_only=False
)

csv_logger = tf.keras.callbacks.CSVLogger(os.path.join(WORK_DIR, "training_log.csv"))
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
```

```python
import math
import numpy as np

# parameters you can tune
IMAGES_PER_SHARD = 5000      # how many images to use per shard (tune to memory & runtime)
EPOCHS_PER_SHARD = 1          # typically 1; do 1 pass per shard
FULL_PASSES = 2               # how many full passes over entire training dataset you want
VAL_EVAL_AT = 1               # run val evaluation after each shard? (we will do val after each shard)

n_train = len(train_df)
num_shards = math.ceil(n_train / IMAGES_PER_SHARD)
print("Train size:", n_train, "Shards:", num_shards)
```

```python
start_shard = 0
# Optionally resume from last checkpoint by loading weights (if exists)
latest = tf.train.latest_checkpoint(ckpt_dir)
if latest:
    print("Resuming from:", latest)
    model = tf.keras.models.load_model(latest)
else:
    print("No checkpoint found  starting fresh")

for pass_i in range(FULL_PASSES):
    print(f"\n=== FULL PASS {pass_i+1}/{FULL_PASSES} ===")
    # optional: shuffle the training dataframe order across shards
    train_df = train_df.sample(frac=1, random_state=pass_i).reset_index(drop=True)
    for s in range(start_shard, num_shards):
        lo = s * IMAGES_PER_SHARD
        hi = min((s+1) * IMAGES_PER_SHARD, n_train)
        shard_df = train_df.iloc[lo:hi]
        print(f"\n--- Training shard {s+1}/{num_shards} images {lo}:{hi} size={len(shard_df)} ---")

        train_ds_shard = make_dataset_from_df(shard_df, shuffle=True)
        val_ds_full = make_dataset_from_df(val_df, shuffle=False)

        history = model.fit(
            train_ds_shard,
            validation_data=val_ds_full,
            epochs=EPOCHS_PER_SHARD,
            callbacks=[checkpoint_callback, best_callback, csv_logger, reduce_lr, early_stop],
            verbose=1
        )

        # After each shard's training we can clear GPU memory caches (optional)
        import gc
        gc.collect()
    # end shard loop
# end full pass loop
```
